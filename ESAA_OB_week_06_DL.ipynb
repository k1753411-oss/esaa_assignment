# 4장. 딥러닝의 시작
### 4.1. 인공 신경망의 한계와 딥러닝 출현

퍼셉트론은 다수의 신호(흐름이 있는)를 입력으로 받아 하나의 신호를 출력. 이 신호를 입력받아 '흐른다/안 흐른다(1/0)' 정보를 앞으로 전달하는 원리로 작동.
* AND 게이트 : 모든 입력이 1일 때 작동. 어떤 하나라도 0을 가지면 작동 중지.
* OR 게이트 : 입력에서 둘 중 하나만 1이거나 둘 다 1일 때 작동. 입력 모두가 0을 갖는 경우를 제외한 나머지가 모두 1을 가짐.
* XOR 게이트 : 배타적 논리합. 입력 두 개 중 한 개만 1일 때 작동.

데이터가 비선형적으로 분리되어 제대로된 분류 어려움. 단층 퍼셉트론에서는 AND, OR 연산에 대해서는 학습이 가능하지만, XOR에 대해서는 학습 불가능.

↓

SOL) 입력층과 출력층 사이에 하나 이상의 중간층(은닉층)을 두어 비선형적으로 분리되는 데이터도 학습이 가능하도록 다층 퍼셉트론(MLP)를 고안.

심층 신경망(DNN): 은닉층이 여러 개 있는 신경망 = 딥러닝.

### 4.2. 딥러닝 구조

**딥러닝**: 여러 층을 가진 입공 신경망을 사용하여 학습을 수행하는 것.

#### **1. 딥러닝 용어 및 구성요소**
* 입력층(input layer): 데이터를 받아들이는 층.
* 은닉층(hidden layer): 모든 입력 노드부터 입력 값을 받아 가중합을 계산하고, 이 값을 활성화 함수에 적용하여 출력층에 전달하는 층.
* 출력층(output layer): 신경망의 최종 결괏값이 포함된 층.

* 가중치(weight): 노드와 노드 간 연결 강도. 입력 값이 연산 결과에 미치는 영향력을 조절.
* 바이어스(bias): 가중합을 더해 주는 상수로, 하나의 뉴런에서 활성화 함수를 거쳐 최종적으로 출력되는 값을 조절하는 역할을 함.
* 가중합(weighted sum), 전달 함수: 가중치와 신호의 곱을 합한 것.
* 활성화 함수(activation function): 신호를 입력받아 이를 적절히 처리하여 출력해주는 함수. 전달 함수에서 받은 값을 출력할 때 일정 기준에 따라 출력 값을 변화시키는 비선형 함수.(sigmoid, tanh, ReLU 등)
  * sigmoid function: 선형 함수의 결과를 0~1 사이에서 비선형 형태로 변형. 주로 분류 문제를 확률적으로 표현. But, DL의 깊이가 깊어지면 vanishing gradient 문제가 발생하여 DL 모델에서는 잘 사용하지 않음.
  * tanh function: 선형 함수의 결과를 -1~1 사이에서 비선형 형태로 변형. 여전히 vanishing gradient 문제가 발생.
  * ReLU function: 입력이 음수일 때는 0 출력, 양수일 대는 x 출력. 경사하강법에 영향을 주지 않아 학습 속도 빠름.  vanishing gradient 문제 발생x. 일반적으로 은닉층에서 사용됨. But, 음수 값을 받으면 항상 0 출력하여 학습 능력 감소.-> Leaky ReLU
  * Leaky ReLU: 입력 값이 음수면 0이 아닌 0.001처럼 매우 작은 수 반환.
  * Softmax function: 입력 값을 0~1 사이에 출력되도록 정규화하여 출력 값들의 총합이 항상 1이 되도록. DL에서 출력 녿의 활성화 함수로 많이 사용.
* 손실 함수(loss function): 가중치 학습을 위해 출력 함수의 결과와 실제 값 간의 오차를 측정하는 함수.
  * 평균 제곱 오차(MSE): 실제 값과 예측 값의 차이를 제곱하여 평균. 회귀에서 손실 함수로 주로 사용됨.
  * Cross-entropy error(CEE): 분류 문제에서 원-핫 인코딩 했을 때 사용할 수 있는 오차 계산 법. 두 개의 확률 분포 차이를 이용하여 시그모이드의 영향을 덜 받음. MSE보다 빠름.

#### **2. 딥러닝 학습**
: 순전파, 역전파

* 순전파(feedforward): 네트워크에 훈련 데이터가 들어올 때 발생. 데이터를 기반으로 예측 값을 계산하기 위해 전체 신경망을 교차해 지나감. 즉, 모든 뉴런이 이전 층의 뉴런에서 수신한 정보에 변환(가중합 및 활성화 함수)를 적용하여 다음 층(은닉층)의 뉴런으로 전송하는 방식. 네트워크를 통해 입력 데이터를 전달하며, 데이터가 모든 층을 통과하고 모든 뉴런이 계산을 완료하면, 그 예측 값은 최종 층(출력층)에 도달.
* 역전파(backpropagation): 손실 함수로 네트워크의 손실, 오차를 추정. 손실 함수 비용이 0에 가깝도록 모델이 훈련을 반복하면서 가중치 조정. 손실이 계산되면 그 정보는 역으로 전파(출력층->은닉층->입력층)됨. 모든 뉴런에 대해 진행하여 계산된 각 뉴런 결과를 또다시 순전파의 가중치 값으로 사용.

#### **3. 딥러닝의 문제점과 해결 방안**
딥러닝의 핵심은 활성화 함수가 적용된 여러 은닉층을 결합하여 비선형 영역을 표현하는 것. 은닉층 개수가 많을수록 데이터 분류가 잘됨.

은닉층 많을 때의 문제
* 과적합 문제 -> dropout(학습 과정 중 임의로 일부 노드를 학습에서 제외.)
* vanishing gradient problem: 은닉층이 많은 신경망에서 주로 발생. 출력층에서 은닉층으로 전달되는 오차가 크게 줄어들어 학습되지 않는 현상. -> ReLU 활성화 함수 사용.
* 성능이 나빠지는 문제 발생 -> 확률적 경사 하강법 , 미니 배치 경사 하강법
  * 배치 경사 하강법(BGD): 전체 데이터셋에 대한 오류를 구한 후 기울기를 한 번만 계산하여 모델의 파라미터 업데이트. 즉, 전체 훈련 데이터셋에 대해 가중치를 편미분. 확습 오래 걸림.
  * 확률적 경사 하강법(SGD): 임의로 선택한 데이터에 대해 기울기를 계산하는 방법. 적은 데이터 사용하여 계산 빠름. BGD보다 정확도가 낮을 수 있지만 속도 빠름.
  * 미니 배치 경사 하강법: 전테 데이터셋을 미니 배치 여러 개로 나누고, 미니 배치 한 개마다 기울기를 구한 후 그것의 평균 기울기를 이용하여 모델을 업데트해서 학습. 안정적이면서 속도 빠름. 가장 많이 사용.

**옵티마이저**: 확률적 경사 하강법의 파라미터 변경 폭이 불안정한 문제를 해결하기 위해 학습 속도와 운동량 조절.
* 속도 조정 방법
  * Adagrad(Adaptive gradient): 업데이트 횟수에 따라 학습률 조정. 많이 변화하지 않는 변수들의 학습률을 크게 하고, 많이 변화하는 변수들의 학습률은 작게 함. 즉, 많이 변화한 변수는 최적 값에 근접했을 것이라는 가정하에 작은 크기로 이동하면서 세밀하게 값을 조정. 파라미터마다 다른 학습률 주기 위해 G함수(G값은 이전 G 값이 누적) 추가. / 기울기가 0에 수렴하는 문제가 있어 사용x. 대신 알엠에스프롭 사용.
  * Adadelta(Adaptive delta): 아다그라드에서 G 값이 커짐에 따라 학습이 멈추는 문제를 해결. 아다그라드 수식에 학습률을 D함수(가중치의 변화량 크기를 누적한 값)로 변환. 학습률에 대한 하이퍼파라미터 필요x.
  * RMSProp: 아다그라드의 G값이 무한히 커지는 것 방지. G함수에 감마만 추가.
* 운동량 조정 방법
  * 모멘텀(Momentum): 경사하강법과 마찬가지로 매번 기울기를 구하지만, 가중치를 수정하기 전에 이전 수정 방향(+/-)을 참고하여 같은 방향으로 일정한 비율만 수정. 지그재그 현상 감소. like 관성. SGB와 함께 사용됨.
  * 네스테로프 모멘텀(Nesterov Accelerated Gradient, NAG): 모멘텀 값과 기울기 값이 더해져 실제 값을 만드는 기존 모멘텀과 달리 모멘텀 값이 적용된 지점에서 기울기 값을 계산. 모멘텀으로 절반 정도 이동한 후 어떤 방식으로 이동해야 하는지 다시 계산하여 결정.
* 속도와 운동량에 대한 혼용 방법
  * Adam(Adaptive Moment Estimation): 모멘텀과 RMSProp의 장점을 결합한 경사 하강법. RMSProp의 특징인 기울기의 제곱을 지수 평균한 값과 모멘텀 특징인 v(i)를 수식에 활용. 즉, RMSProp의 G함수와 모멘텀의 v(i)를 사용하여 가중치 업데이트.

#### **4. 딥러닝 사용 시 이점**
* 특성 추출: 컴퓨터가 입력받은 데이터를 분석하여 일정한 패턴이나 규칙을 찾아내려면 사람이 인지하는 데이터를 컴퓨터가 인지할 수 있는 데이터로 변환해 주어야 하는데, 이때 데이터별로 어떤 특징을 가지고 있는지 찾아내고, 그것을 토대로 데이터를 벡터로 변환하는 작업.
* 빅데이터의 효율적 활용: 딥러닝은 무조건 데이터 많아야 해.

```python
import torch
```

렐루 함수와 소프트 맥스 함수

```python
class Net(torch.nn.Module):
  def __init__(self, n_feature, n_hidden, n_output):
    super(Net, self).__init__()
    self.hidden = torch.nn.Linear(n_feature, n_hidden) # 은닉층
    self.relu = torch.nn.ReLu(inplace=True)
    self.out = torch.nn.Linear(n_hidden, n_output) # 출력층
    self.softmax = torch.nn.Softmax(dim=n_output)
  def forward(self, x):
    x = self.hidden(x)
    x = self.relu(x) # 은닉층을 위한 렐루 활성화 함수
    x = self.out(x)
    x = self.softmax(x) # 출력층을 위한 소프트맥스 활성화 함수
    return x
```

# 답러닝 알고리즘
## **심층 신경망**
* DNN은 입력층과 출력층 사이에 다수의 은닉층을 포함하는 인공 신경망.
* 다수의 은닉층을 추가하여 별도의 트릭없이 비선형 분류 가능.
* 학습을 위한 연산량이 많고 기울기 소멸 문제 등이 발생 가능.

## **합성곱 신경망**
* CNN은 합성곱층과 풀링층을 포함하는 이미지 처리 성능이 좋은 인공 신경망 알고리즘.
* 영상 및 사진이 포함된 이미지 데이터에서 객체를 탐지하거나 객체 위치를 찾아내는데 유용한 신경망.
* 대표적인 합성곱 신경망으로 LeNet-5와 AlexNet, 층을 더 깊게 쌓은 신경망으로는 VGG, GoogLeNet, ResNet 등.
* 각 층의 입출력 형상을 유지
* 이미지의 공간 정보를 유지하면서 인접 이미지와 차이가 있는 특징을 효과적으로 인식.
* 복수 필터로 이미지의 특징을 추출하고 학습.
* 추출한 이미지의 특징을 모으고 강화하는 풀링층 존재.

## **순환 신경망**
* 시계열 데이터 같은 시간 흐름에 따라 변화하는 데이터를 학습시키기 위한 인공 신경망.
* 시간성을 가진 데이터 많음.
* 시간성 정보를 이용하여 데이터의 특징을 잘 다룸.
* 데이터 동적, 길이 가변적.

## **제한된 볼츠만 머신**
* 가시층, 은닉층으로 구성된 모델.
* 가시층은 은닉층과만 연결됨.

## **심층 신뢰 신경망**
*  입력층과 은닉층으로 구성된 제한된 볼츠만 머신을 블록처럼 여러 층으로 쌓은 형태로 연결된 신경망.
* 사전 훈련된 제한된 볼츠만 머신을 층층이 쌓아 올린 구조로 레이블이 없는 데이터에 대한 비지도 학습이 가능.
* 순차적으로 심층 신뢰 신경망을 학습시켜 가면서 계층적 구조를 생성.
* 위로 올라갈수록 추상적 특징 추출.
* 학습된 가중치를 다층 퍼셉트론의 가중치 초깃값으로 사용.
