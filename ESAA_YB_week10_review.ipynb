# **수상작 리뷰**

##데이터셋 정보
주제 : 은행 데이터셋을 사용한 이진 분류
목표 : 고객이 은행 정기예금에 가입할지 여부를 예측

변수 : id, age, job, marital, education, balance, housing, loan, contact


## 전처리

1. 데이터 로드 및 타입 설정 : 범주형 변수(CATS)는 모두 category 타입으로 변환해서 메모리 절약 + 인코딩 효율 향상

2. 수치형 변수의 주기/로그 변환 생성 :
* _duration_sin, _duration_cos : 통화 duration을 540초(=9분) 주기로 변환 → 시간의 주기성 반영
* _balance_log : balance를 로그 스케일로 변환 → 극단값 완화
* _balance_sin, _balance_cos : 잔액의 패턴을 주기적으로 표현
* _age_sin, _pdays_sin : 나이(age)와 지난 연락일(pdays)에 대한 주기적 패턴 반영

3. 수치형 변수 간 상호작용(곱) 생성 :모든 수치형 변수 2개 조합의 곱(x1 * x2)을 생성.

4. 수치형+범주형 조합 feature 생성 : 수치형 + 범주형 조합, 범주형 + 범주형 조합 등 모든 2변수 조합의 문자열 feature 생성.

5. factorization + Target Encoding : orig 데이터에서 각 변수별로 y의 평균값(= 그 범주의 타겟 확률)을 계산.

6. 원본 데이터(orig)의 통계 기반 인코딩(TE_ORIG, CE_ORIG) : 각 값이 원본 데이터에서 얼마나 등장했는지 → 등장 횟수로 인코딩. 즉, 빈도 인코딩.

## 모델링
1. 데이터 분할 (Stratified K-Fold) : 데이터를 라벨 분포가 동일하게 유지되도록 5개의 fold로 분할.

2. 교차검증 내 Target Encoding : 교차검증 내에서 훈련 데이터 기준으로만 Target Encoding을 수행함.

3. 모델 설정 및 학습 (TabM_D_Classifier) : 딥러닝 기반 tabular classifier인 TabM_D_Classifier 사용.

4. 예측 및 앙상블 평균 : 각 fold에서 모델 학습, 검증 데이터 예측 → oof (out-of-fold), 테스트 예측은 fold별로 저장해둔 뒤 나중에 평균.

## 차별점 및 배울점
딥러닝 기반 tabular classifier을 사용하여
LightGBM·XGBoost보다 계산은 무겁지만, feature 간 상호작용을 학습적으로 찾아내는 게 장점을 활용함이 인상깊었다.
다양한 OOF를 여러 개 생성하고(동일한 모델과 다른 하이퍼 매개변수/FE가 아닌 다른 모델을 사용하는 것이 이상적임) 천천히 앙상블에 추가하는 것이 수상에 중요한 역할을 함을 배울 수 있었다.
