# 5장. 합성곱 신경망

### **[CNN]**
기존 : 순전파 과정에 따라 계산된 오차 정보가 신경망의 모든 노드(출력층->은닉층->입력층)으로 전송. ==> 계산 과정 복잡, 많은 자원 요구, 계산 시간 ↑

  ↓

SOL) **CNN**: 이미지 전체를 한 번에 계산X. 이미지의 국소적 부분을 계산함으로써 시간과 자원 절약, 이미지의 세밀한 부분까지 분석 가능.

---
이미지 분석: nxn 배열을 flattening하고, 각 픽셀에 가중치를 곱하여 은닉층에 전달. -> 데이터의 공간적 구조 무시. -> SOL) 합성곱층

**합성곱 신경망(Convolutional Neural Network, CNN or ConvNet)** : 음성 인식이나 이미지/영상 인식에서 주로 사용됨. 다차원 배열 처리에 특화. 계층 5개로 구성됨.

1. 입력층
* 입력 이미지 데이터(높이, 너비, 채널의 값을 갖는 3차원 데이터)가 최초로 거치게 되는 계층.
* 채널은 이미지가 gray scale이면 1, RGB(컬러)면 3 값을 가짐.
2. 합성곱층
* 층과 층을 이어주는 역할.
* 입력 데이터에서 특성을 추출.
* 입력 이미지가 들어왔을 때 이미지에 대한 특성을 감지하기 위해 **커널이나 필터** 사용.
* 커널/필터는 이미지의 모든 영역을 훑으면서 특성 추출. 추출된 결과물이 **feature map**.
* 커널은 3x3, 5x5 크기로 적용되는 것이 일반적.
* stride라는 지정된 간격에 따라 순차적으로 이동.
* 하이퍼파라미터: 필터 개수(K), 필터 크기(F), 스트라이드(S), 패딩(P)
* 출력 데이터
  * W2 = (W1 - F + 2P)/S +1
  * H2 = (H1 - F + 2P)/S +1
  * D2 = K
3. 풀링층
* 특성 맵의 차원을 다운 샘플링하여 연산량 감소시킴.
* 최대 풀링(max pooling): 대상 영역에서 최댓값을 추출. 주로 많이 사용됨.
* 평균 풀링(average pooling): 대상 영역에서 평균을 반환.
* 하이퍼파라미터: 필터 크기(F), 스트라이드(S)
* 출력 데이터
  * W2 = (W1 - F)/S +1
  * H2 = (H1 - F)/S +1
  * D2 = D1
4. 완전연결층(fully connected layer)
* fc로 들어가기 위해 3차원 벡터에서 1차원 벡터로 펼쳐짐(flatten).
* fc층은 펼쳐진 데이터에 가중치(선)를 곱해서 다음 단계로 정보를 '변환'.
* 앞쪽 뉴런과 뒤쪽 뉴런이 빠짐없이 모두(Fully) 연결(Connected) 되어 있다"는 뜻. 이것은 블록의 종류이지, 위치가 아님.
5. 출력층: softmax func 사용됨. 이미지가 각 레이블에 속할 확률 값이 출력되며, 이때 가장 높은 확률 값을 갖는 레이블이 최종 값으로 선정됨.

---
* 1D CNN: 필터가 시간을 축으로 좌우로만 이동. 출력 형태는 1D 배열이 되며, 그래프 곡선 완화 시 많이 사용.
* 1D CNN: 필터가 방향 두 개로 움직임. 입력(W,H), 필터(k,k), 출력(W,H).
* 3D CNN: 필터가 방향 세 개로 움직임. 입력(W,H,L), 필터(k,k,d), 출력(W,H,L).  d<L 유지!
* 3D 입력을 갖는 2D CNN: LeNet, VGG
* 1X1 CNN: GoogLeNet
